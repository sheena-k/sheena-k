```python
q3 = np.percentile(data['spirit_servings'], 75,method="midpoint")
```

```python

print(q2)
print(q3)
```

```python

iqr = q3 - q1
```

```python

low_limit = q1 - 1.5*iqr
up_limit = q3 + 1.5*iqr
     
```

```python

print(low_limit)
print(up_limit)
```

```python
outlier =[]
for x in data['spirit_servings']:
  if((x>up_limit) or (x<low_limit)):
     
        outlier.append(x)
     
```
