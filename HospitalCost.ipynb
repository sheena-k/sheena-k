```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("train.csv")
```

```python
data.head(10)
```

```python
data.describe(include="all")
```

```python
data=data.drop(["Race", "Gender", "Facility Id","Unnamed: 0"],axis=1)
```

```python
correlation_matrix_with_target = data.corr()

```

```python
correlation_with_target = correlation_matrix_with_target["Total Costs"].drop("Total Costs")
```

```python
sorted_correlation_with_target = correlation_with_target.sort_values(ascending=True)
```

```python
plt.figure(figsize=(25,10))
sns.heatmap(correlation_matrix_with_target, annot=True,fmt='.2')
plt.title('Correlation Matrix Including Target Variable')
plt.show()
```

```python
plt.figure(figsize=(10, 8))
sns.barplot(x=sorted_correlation_with_target.values, y=sorted_correlation_with_target.index,legend =False)
plt.xlabel('Correlation with Target')
plt.ylabel('Features')
plt.title('Feature Correlation with Target Variable')
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
X = data.drop("Total Costs",axis=1)
Y = data["Total Costs"].values.ravel()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
```

```python
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
#Extract and Plot Feature Importances (Coefficients)
feature_importances = pd.Series(model.coef_, index=X.columns).sort_values(ascending=False)
```

```python
# Plot Feature Importances from Linear Regression
plt.figure(figsize=(10, 8))
sns.barplot(x=feature_importances.values, y=feature_importances.index)
plt.xlabel('Linear Regression Coefficient')
plt.ylabel('Features')
plt.title('Feature Importance Based on Linear Regression Coefficients')
plt.show()
```
